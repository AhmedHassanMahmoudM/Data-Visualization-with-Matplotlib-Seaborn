# Distribution Plots

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

* 1) Univariate Plots
    * 1. distplot
    * 2. kdeplot
* 2) Bivariate Plots
    * 1. jointplot
    * 2. pairplot

## Imports

```python
import seaborn as sns

%matplotlib inline
sns.set(rc={'figure.figsize': [8, 8]}, font_scale=1.2)
```

## 1) Univariate Plots
plots that deal with one variable
* 1. distplot
* 2. kdeplot

### 1. distplot

The distplot shows the distribution of a univariate set of observations.

```python
sns.distplot(tips['total_bill'])
```

Change the color a little bit, valid options for colors: r, g, b, c, m, y, k

```python
sns.distplot(tips['total_bill'], color='m')
```

To remove the kde layer and just have the histogram use:

```python
sns.distplot(tips['total_bill'], kde=False, color='m')
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30, color='m')
```

```python
sns.distplot(tips['tip'], kde=False, bins=30, color='m')
```

```python
sns.distplot(tips['size'], kde=False, color='m')
```

### 2. kdeplot

kdeplots are [Kernel Density Estimation plots](http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth). These KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value. For example:

```python
sns.kdeplot(tips['total_bill'], color='m')
```

```python
sns.kdeplot(tips['total_bill'], shade=True, color='m')
```

```python
sns.kdeplot(tips['tip'], shade=True, color='m')
```

```python
sns.kdeplot(tips['size'], shade=True, color='m')
```

## 2) Bivariate Plots
plots that deal with two variables
* 1. jointplot
* 2. pairplot

### 1. jointplot

jointplot() allows you to basically match up two distplots for bivariate data. With your choice of what **kind** parameter to compare with: 
* “scatter” 
* “reg” 
* “kde” 
* “hex”

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
```

Change the size and the color a little bit

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter', height=8, color='m')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg', height=8, color='m')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde', height=8, color='m')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex', height=8, color='m')
```

```python
# try some cool colors
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde', cmap=cmap, height=8)
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex', cmap=cmap, height=8)
```

### 2. pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns). 

```python
sns.pairplot(tips)
```

Change the size a little bit

```python
sns.pairplot(tips, height=3.5, aspect=1.3)
```

Choose custom columns

```python
sns.pairplot(tips, vars=['total_bill', 'tip'], height=5, aspect=1.3)
```

Use hue to show graph based on the hue category values

```python
sns.pairplot(tips, hue='sex', height=3, aspect=1.3)
```

```python
sns.pairplot(tips, hue='day', height=3, aspect=1.3)
```

```python
sns.pairplot(tips, hue='time', height=3, aspect=1.3)
```

**Customization of pairplot** <br><br>
basically pairplot is based on PairGrid so if we want to make customizations we should use PairGrid

```python
g = sns.PairGrid(iris, height=3, aspect=1.3)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot)
```

```python
g = sns.PairGrid(iris, height=3, aspect=1.3)
g.map_diag(sns.kdeplot, shade=True)
g.map_offdiag(sns.kdeplot, shade=True)
```

```python
# try some cool colors
cmap = sns.cubehelix_palette(n_colors=8, as_cmap=True, dark=0, light=1, reverse=True)

g = sns.PairGrid(iris, height=3, aspect=1.3)
g.map_diag(sns.kdeplot, shade=True, color='m')
g.map_offdiag(sns.kdeplot, shade=True, cmap=cmap)
```
