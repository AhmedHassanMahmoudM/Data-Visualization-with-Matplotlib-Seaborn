Let's go through examples of each!

```python
import seaborn as sns

%matplotlib inline
sns.set(rc={'figure.figsize': [7, 7]}, font_scale=1.2)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## 1) Categorical distribution plots
* 1. boxplot
* 2. violinplot

boxplots and violinplots are used to shown the distribution of categorical data.

### 1. boxplot

A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
sns.boxplot(x="day", y="total_bill", data= tips)
```

```python
sns.boxplot(x="total_bill", y="day", data=tips)
```

change the color palette

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette='viridis')
```

```python
sns.boxplot(x="time", y="total_bill", data=tips)
```

```python
sns.boxplot(x="smoker", y="total_bill", data=tips)
```

```python
sns.boxplot(x="sex", y="total_bill", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", hue="sex",data=tips, palette='Set3')
```

```python
sns.boxplot(x="time", y="total_bill", hue="smoker",data=tips)
```

```python
sns.boxplot(x="sex", y="total_bill", hue="smoker",data=tips)
```

```python
sns.boxplot(x="sex", y="tip", hue="time",data=tips)
```

### 2. violinplot
A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="total_bill", y="day", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, palette='viridis')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue='sex', palette='viridis')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue='sex', split=True, palette='viridis')
```

## 2) Categorical scatter plots
* 1. stripplot
* 2. swarmplot

### 1. stripplot
The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue='smoker')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue='sex')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue='time')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue='sex', dodge=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue='time', dodge=True, palette='viridis')
```

**we can mix stripplot with another plots like boxplot and violinplot to get better representation**

```python
sns.boxplot(x="day", y="total_bill", data=tips)
sns.stripplot(x="day", y="total_bill", color="#474646", size=5, data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
sns.stripplot(x="day", y="total_bill", color="#474646", size=5, data=tips)
```

### 2. swarmplot

swarmplot draw a categorical scatterplot with non-overlapping points. This function is similar to stripplot() , but the points are adjusted (only along the categorical axis) so that they don't overlap.
It can give a better representation of the distribution of observations, although it only works well for relatively small datasets.

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue='smoker')
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue='sex')
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue='time')
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue='sex', dodge=True)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue='time', dodge=True, palette='rocket')
```

**we can mix swarmplot with another plots like boxplot and violinplot to get better representation**

```python
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", color="#474646", size=5, data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", color="#474646", size=5, data=tips);
```

## 3) Categorical estimate plots

* 1. barplot
* 2. countplot

These very similar plots allow you to get aggregate data off a categorical feature in your data.

### 1. barplot
is a general plot that allows you to aggregate the categorical data based off some function, by default the mean:

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.barplot(x='time',y='total_bill', data=tips)
```

```python
sns.barplot(x='smoker', y='tip', data=tips)
```

You can change the estimator object to your own function, that converts a vector to a scalar:

```python
import numpy as np
```

```python
# sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.mean)
# sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
# sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.median)
# sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.max)
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.min)
# sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.sum)
```

```python
sns.barplot(x='smoker', y='total_bill',data=tips, estimator=np.sum)
```

```python
sns.barplot(x='day', y='total_bill',data=tips, estimator=np.median)
```

### 2. countplot

This is essentially the same as barplot except the estimator is explicitly counting the number of occurrences. Which is why we only pass the x value:

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.countplot(y='sex', data=tips)
```

```python
sns.countplot(x='smoker', data=tips)
```

```python
sns.countplot(x='day', data=tips, palette='rocket')
```

```python
sns.countplot(x='time', data=tips, hue='day', palette='rocket')
```

## 4) Generalization of catplot
* 1. generate all plots with catplot
* 2. multible plots

### 1. generate all plots with catplot

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips, height=6, aspect=1.3)
```

```python
sns.catplot(x="day", y="total_bill", data=tips, height=6, aspect=1.3, kind="violin", hue='smoker', split=True)
```

```python
sns.catplot(x="day", y="total_bill", data=tips, height=6, aspect=1.3, kind="bar", hue='time', estimator=np.sum)
```

```python
sns.catplot(x="day", data=tips, height=6, aspect=1.3, kind="count", hue='sex')
```

```python
sns.catplot(x="day", y="total_bill", data=tips, height=6, aspect=1.3, kind="strip", hue='time', palette='rocket')
```

```python
sns.catplot(x="day", y="total_bill", data=tips, height=6, aspect=1.3, kind="swarm", hue='time', palette='rocket')
```

## 2. multiple plots
you can make multiple plots from specific variable values using **col, row** attributes in catplot

```python
sns.catplot(x="day", y="total_bill", data=tips, hue="smoker", height=5, aspect=1.2, kind="box", col="time")
```

```python
sns.catplot(x="day", y="total_bill", data=tips, height=5, aspect=1.2,
            hue="sex", kind="violin", split=True, col="time", row="smoker")
```

```python
sns.catplot(x="day", y="total_bill", data=tips, height=5, aspect=1.2, palette='Set1',
            hue="sex", kind="swarm", col="time", row="smoker")
```
